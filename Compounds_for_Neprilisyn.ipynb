Importing Libraries

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

**Target search for skin **



```python
target = new_client.target
target_query = target.search('skin')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[10]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
df
```

```python
df.type.unique()
```

## ***Handling missing data***

```python
df2 = df[df.standard_value.notna()]
df2 = df[df.value.notna()]
df2
```

```python
df2['standard_value']
```

```python
print(df2['value'].values)
```

##Data pre_processing

```python
bioactivity = []
for x in df2.standard_value:
  if float(x) >= 10000:
    bioactivity.append("inactive")
  elif float(x) <= 1000:
    bioactivity.append("active")
  else:
    bioactivity.append("Intermediate")
```

```python
print(len(bioactivity))
```

***Creating and saving 'chembl_id','canonical_smiles','standard_value', 'bioactivity', into lists and concatenating them into a new dataframe***

```python
mol_chembl_id= []
for x in df2.molecule_chembl_id:
  mol_chembl_id.append(x)
```

```python
canonical_smiles = []
for x in df2.canonical_smiles:
  canonical_smiles.append(x)
```

```python
standard_value = []
for x in df2.standard_value:
  standard_value.append(x)
```

```python
#combine the 4 lists
data_tuples = list(zip(mol_chembl_id, canonical_smiles, bioactivity, standard_value))
df3 = pd.DataFrame( data_tuples,  columns=['mol_chembl_id', 'canonical_smiles', 'bioactivity', 'standard_value'])
df3
```
